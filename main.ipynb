````python
import requests
from bs4 import BeautifulSoup

url = 'https://autos.yahoo.com.tw/popular-bikes'
html = requests.get(url)

bs = BeautifulSoup(html.text, 'lxml')
````

````python
import pandas as pd
import os
import urllib

path = '/Users/rich/Desktop/FJU_Programming_Design/static/images'

    
# 找出所有需要的機車
data = bs.find_all('div', class_ = 'popular-highlight')
rows = []
for i in range(len(data)):
    bike_type = data[i].find('span')
    print(bike_type.text) 
    bikes = data[i].find_all('div', class_ = 'ranking-wrapper')
    for bike in bikes:

        # 機車資訊
        info = bike.find_all('span')
        
        # 機車排名
        rank = info[0].text
        print('排名', rank, end = ' ')

        # 機車價位
        price = info[1].text
        print(price, '萬', end = ' ')

        # 機車名稱
        name = info[2].text
        print(name)
        print('')

        # 下載照片
        all_links = bike.find_all('img')
        for link in all_links:
            src = link.get('src')
            attrs = [src]
            for attr in attrs:
                if attr is not None and ('.jpeg' in attr or '.png' in attr):
                    full_path = attr
                    file_name = os.path.join(path, f"{name}.png")
                    image = urllib.request.urlopen(full_path)
                    f = open(file_name, 'wb')
                    f.write(image.read())
                    f.close()
                    
                        
        rows.append([bike_type, name ,rank, price])
columns = ['type', 'name', 'ranked', 'price']

df = pd.DataFrame(rows, columns = columns)
````

````python
df
````

````python
df.type = df.type.astype(str)

for i in range(len(df)):
    df.type[i] = df.type[i].replace('<span>', '').replace('</span>', '')

df.ranked = df.ranked.astype(int)
df.price = df.price.astype(float)

ranked_mapping = {1: 1, 2: 2, 3: 3, 4: 1, 5: 2}
df['ranked'] = df['ranked'].replace(ranked_mapping)
````

````python
# 串接openai api問chatgpt覺得貴不貴
import openai
from dotenv import dotenv_values
config = dotenv_values(".env")
config['OPENAI_API_KEY']
openai.api_key = config['OPENAI_API_KEY']

def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0, 
    )
    return response.choices[0].message["content"]
````

````python
# 連接資料庫
import pyodbc
conn = pyodbc.connect(
    'Driver={ODBC Driver 17 for SQL Server};' #驅動程式
    'Server=小黑;' #伺服器
    'Database=pythondb;' # 資料庫
    'Trusted_Connection=yes;' # Windows驗證
)
cursor = conn.cursor()
````

````python
# 建立資料表
sql = '''
CREATE TABLE Bikes(
    id INT IDENTITY(1,1) PRIMARY KEY,
    type VARCHAR(50),
    name VARCHAR(50),
    ranked INT,
    price FLOAT,
    cp_value INT,
    appearance INT
)
'''

conn.execute(sql)  
conn.commit()
````

````python
# 使用df.iterrows()進行迭代
for index, row in df.iterrows():
    # 取得資料行的值
    bike_type = row['type']
    name = row['name']
    ranked = row['ranked']
    price = row['price']
    cp_value = row['cp_value']
    appearance = row['appearance']

    # 插入資料
    sql = 'INSERT INTO Bikes (type, name, ranked, price, cp_value, appearance) VALUES (?, ?, ?, ?, ?, ?)'
    cursor.execute(sql, (bike_type, name, ranked, price, cp_value, appearance))
conn.commit()
conn.close()
````

````python
import os
import matplotlib.pyplot as plt

labels = ['Ranked', 'CP Value', 'Appearance']

save_directory = '/Users/rich/Desktop/FJU_Programming_Design/static/radarChart'
os.makedirs(save_directory, exist_ok=True)

for i, row in df.iterrows():
    values = row[['ranked', 'cp_value', 'appearance']].values.tolist()
    values += values[:1]
    angles = [n / float(len(labels)) * 2 * 3.14159 for n in range(len(labels))]
    angles += angles[:1]

    fig, ax = plt.subplots(figsize=(6, 6), subplot_kw={'polar': True})
    ax.plot(angles, values, linewidth=1, linestyle='solid')
    ax.fill(angles, values, alpha=0.25)
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels)
    ax.set_ylim(0, 4)

    filename = f"{row['name']}.png"
    save_path = os.path.join(save_directory, filename)

    plt.savefig(save_path)
    plt.close()
````
